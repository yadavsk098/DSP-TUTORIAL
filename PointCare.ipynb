```python
from rpeakdetect import detect_beats
```

```python
# array hai rr peaks ki
rr1 = detect_beats(ecg1[0], 300) # rr peaks for normal signal
rr2 = detect_beats(ecg2[0], 300) # rr peaks for AF signal
rr3 = detect_beats(ecg3[0], 300) # rr peaks for Noisy signal
rr4 = detect_beats(ecg4[0], 300) # rr peaks for Other signal
```

```python
pr1 = np.diff(rr1)
pr2 = np.diff(rr2)
pr3 = np.diff(rr3)
pr4 = np.diff(rr4)
```
